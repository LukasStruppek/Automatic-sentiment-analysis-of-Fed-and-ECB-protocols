```python
import requests
response = requests.get("https://www.finanzen.net/realtimekurse")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find_all(class_="table")[0].select("tr")
lines = [t for t in table[1:]]
```

```python
names = []
kurse = []
schlusskurse = []
change = []
percent = []
time = []
for line in lines:
    names.append(line.select("td")[0].get_text())
    kurse.append(line.select("td")[1].get_text())
    schlusskurse.append(line.select("td")[2].get_text())
    change.append(line.select("td")[3].get_text())
    percent.append(line.select("td")[4].get_text())
    time.append(line.select("td")[5].get_text())
```
